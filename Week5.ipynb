```python
import cv2 as cv 
from google.colab.patches import cv2_imshow 
from skimage import io 
import matplotlib.pyplot as plt 
import numpy as np 

img_f = cv.imread('/content/drive/MyDrive/imagedrive/female.tiff')
img_rgb = cv.cvtColor(img_f,cv.COLOR_BGR2RGB) # konversi channel BGR -> RGB
plt.imshow(img_rgb)
```

```python
#Tugas Praktikum 1
print(' Gamma Correction pada citra ')
print('----------------------------------')
try: 
  gamma = int(input('Masukkan nilai Gamma: '))
except ValueError:
  print('Error, not a number')

original = cv.cvtColor(img_rgb,cv.COLOR_BGR2RGB)
gamma_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
     gamma_image[y,x,c] = np.clip(255 * pow(original[y,x,c]/255,(1/gamma)), 0, 255)
     
final_frame = cv.hconcat((original, gamma_image))
cv2_imshow(final_frame)
```

```python
#Tugas Praktikum 2
original = cv.imread('/content/drive/MyDrive/imagedrive/female.tiff', cv.IMREAD_GRAYSCALE) 
depth_image = np.zeros(original.shape, original.dtype)

print(' Simulasi Image Depth ')
print('----------------------')
try:
  bd = float(input('Masukkan nilai bit depth [1-7]: '))
except ValueError:
  print('Error, not a number')

bit = 2
level = 255 / (pow(2,bit)-1) 

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    depth_image[y,x] = np.clip(round(original[y,x] / level) * level, 0, 255);
     

final_frame = cv.hconcat((original, depth_image))
cv2_imshow(final_frame)
```

```python
#Tugas Praktikum 3
#Jumlah Citra di Average (5)

import glob
from math import log10, sqrt
  
original = cv.imread('/content/drive/MyDrive/Images/galaxy.jpg')
cv_img = []
for img in glob.glob('/content/drive/MyDrive/Images/noises/*.jpg'): 
    n = cv.imread(img)
    cv_img.append(n)

dst = cv_img[0]
noise_number = 5 # nomor (array) gambar pada folder noises
for i in range(noise_number):
    if i == 0:
        pass
    else:
        alpha = 1.0 / (i + 1)
        beta = 1.0 - alpha
        dst = cv.addWeighted(cv_img[i], alpha, dst, beta, 0.0)

def PSNR(original, compressed): 
    mse = np.mean((original - compressed) ** 2) 
    if(mse == 0):  # MSE adalah nol berarti tidak ada noise yang ada dalam sinyal. 
                  # Oleh karena itu PSNR tidak memiliki arti penting.
        return 100
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse)) 
    return psnr,mse

compressed = dst
psnr, mse = PSNR(original, compressed)
print('Nilai MSE adalah',mse,'(dB)')
print('Nilai PSNR adalah',psnr,'(dB)')
galaxy_frame = cv.hconcat((original, dst))
cv2_imshow(galaxy_frame)
```

```python
#Tugas Praktikum 4

print('Image masking (Operator AND)')
print('')

img_org = cv.imread('/content/drive/MyDrive/Images/couple.tiff')

mask = np.zeros(img_org.shape, dtype=np.uint8)
mask = cv.circle(mask, (70,50), 50, (255,255,255), -1)
mask2 = cv.circle(mask, (200,50), 50, (255,255,255), -1)

result = img_org & mask & mask2

org_frame = cv.hconcat((img_org, mask, result))
cv2_imshow(org_frame)
```

```python
#Tugas Praktikum 5

print('Image masking (Operator OR)')
print('')

img_org = cv.imread('/content/drive/MyDrive/Images/couple.tiff')

mask = np.zeros(img_org.shape, dtype=np.uint8)
mask = cv.circle(mask, (70,50), 50, (255,255,255), -1)
mask2 = cv.circle(mask, (200,50), 50, (255,255,255), -1)

result = img_org | mask | mask2

org_frame = cv.hconcat((img_org, mask, result))
cv2_imshow(org_frame)
```

```python
#Tugas Praktikum 5

print('Image masking (Operator NOT)')
print('')

img_org = cv.imread('/content/drive/MyDrive/Images/couple.tiff')

mask = np.zeros(img_org.shape, dtype=np.uint8)
mask = cv.circle(mask, (70,50), 50, (255,255,255), -1)
mask2 = cv.circle(mask, (200,50), 50, (255,255,255), -1)

result = ~img_org & ~mask & ~mask2

org_frame = cv.hconcat((img_org, mask, result))
cv2_imshow(org_frame)
```

```python
#Tugas Praktikum 5

print('Image masking (Operator XOR)')
print('')

img_org = cv.imread('/content/drive/MyDrive/Images/couple.tiff')

mask = np.zeros(img_org.shape, dtype=np.uint8)
mask = cv.circle(mask, (70,50), 50, (255,255,255), -1)
mask2 = cv.circle(mask, (200,50), 50, (255,255,255), -1)

result = img_org ^ mask2

org_frame = cv.hconcat((img_org, mask, result))
cv2_imshow(org_frame)
```

**Tuliskan hasil analisa anda:** 

Operator boolean dalam image masking digunakan untuk memilih atau mengekstraksi bagian tertentu dari gambar berdasarkan nilai pikselnya. Operator ini terdiri dari tiga jenis yaitu AND, OR, dan NOT.

AND: Operator AND digunakan untuk mengambil bagian dari gambar yang memiliki nilai piksel yang sama pada kedua gambar input. Misalnya, jika kita ingin mengambil area di mana gambar 1 dan gambar 2 memiliki nilai piksel yang sama, maka kita dapat menggunakan operator AND. Operator ini berguna untuk menggabungkan dua gambar menjadi satu gambar.

OR: Operator OR digunakan untuk mengambil bagian dari gambar yang memiliki nilai piksel yang sama pada salah satu gambar input atau kedua gambar input. Misalnya, jika kita ingin mengambil area di mana gambar 1 atau gambar 2 memiliki nilai piksel yang sama, maka kita dapat menggunakan operator OR. Operator ini berguna untuk mempertahankan area dari gambar asli, sementara menghapus area yang tidak diinginkan.

NOT: Operator NOT digunakan untuk mengambil bagian dari gambar yang memiliki nilai piksel yang tidak sama pada gambar input. Misalnya, jika kita ingin mengambil area di mana gambar 1 dan gambar 2 tidak memiliki nilai piksel yang sama, maka kita dapat menggunakan operator NOT. Operator ini berguna untuk menghapus area dari gambar yang tidak diinginkan.

Operator boolean dalam image masking dapat digunakan untuk berbagai tujuan seperti menghapus latar belakang, menyeleksi objek tertentu, atau menggabungkan beberapa gambar. Namun, dalam penggunaannya, operator boolean membutuhkan pemahaman yang baik tentang nilai piksel pada gambar dan bagaimana cara pengambilan area yang tepat untuk memastikan hasil akhir yang diinginkan.
